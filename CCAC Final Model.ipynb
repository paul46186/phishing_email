### Test split the dataset

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(newdf["text"], newdf["target"], test_size = 0.25, random_state = 513)
```

### Tokenize the data

```python
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer

lemmatizer = nltk.stem.WordNetLemmatizer()
tokencomp = []
for review in list(train_x):
    tokens = nltk.word_tokenize(str(review))
    lemmatized_token = [lemmatizer.lemmatize(token) for token in tokens if token.isalnum()]
    tokencomp.append([token for token in lemmatized_token if token not in stopwords.words('english')])

comp = []
for review in tokencomp:
    comp.append(" ".join(review))
vectorizer = TfidfVectorizer(ngram_range = (1,2), min_df = 2)
tfidf=vectorizer.fit(comp)

train_x = vectorizer.transform(train_x)
test_x = vectorizer.transform(test_x)
X_train = vectorizer.transform(CCAC['combine'].values.astype('U'))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
Logitmodel = LogisticRegression()

Logitmodel.fit(train_x, train_y)

y_pred_logit = Logitmodel.predict(test_x)
LogProb= Logitmodel.predict_proba(test_x)
LogProb=LogProb[:,1]

y_pred_logit2 = Logitmodel.predict(X_train)
LogProb2= Logitmodel.predict_proba(X_train)
LogProb2=LogProb2[:,1]


acc_logit = accuracy_score(test_y, y_pred_logit)
print("Logit model Accuracy:: {:.2f}%".format(acc_logit*100))
```

### XGBoosting

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
XGmodel = XGBClassifier()
XGmodel.fit(train_x, train_y)
# make predictions for test data
y_predXG = XGmodel.predict(test_x)
XGProb=  XGmodel.predict_proba(test_x)
XGProb=XGProb[:,1]


y_predXG2 = XGmodel.predict(X_train)
XGProb2=  XGmodel.predict_proba(X_train)
XGProb2=XGProb2[:,1]


predictions = [round(value) for value in y_predXG]
# evaluate predictions
accuracy = accuracy_score(test_y, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

### SVC

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.externals import joblib
import pickle
SVMmodel = SVC(probability=True)

SVMmodel.fit(train_x, train_y)

y_pred_SVM = SVMmodel.predict(test_x)
SVMProb=  SVMmodel.predict_proba(test_x)
SVMProb=SVMProb[:,1]

y_pred_SVM2 = SVMmodel.predict(X_train)
SVMProb2=  SVMmodel.predict_proba(X_train)
SVMProb2=SVMProb2[:,1]

acc_SVM = accuracy_score(test_y, y_pred_SVM)
print("SVM model Accuracy: {:.2f}%".format(acc_SVM*100))

# save the model to disk
filename = 'SVM_model.sav'
pickle.dump(SVMmodel, open(filename, 'wb'))
 
```
